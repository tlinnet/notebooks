## Background

ANOVA is analysis of variance between repeated measures of samples.

ANOVA was developed in the 1920’s by R.A. Fisher. F stands for Fischer.

## What is the difference between t-test and F-test ?

Typically a t-test is used to examine the differences between the means of two groups. For example, in an experiment you may want to compare the overall mean for the group on which the manipulation took place vs a control group.

However, if you have more than two groups, you shouldn't just use multiple t-tests as the error adds up (see familywise error) and thus you increase your chances of finding an effect when there really isn't one (i.e. a type 1 error). Therefore when you have more than two groups to compare e.g. in a drugs trial when you have a high dose, low does and a placebo group (so 3 groups), you use ANOVA to examine whether there any differences between the groups.

To examine whether there are differences between the groups you would use the F-ratio, which essentially measures the improvement due to fitting the model i.e. the group means versus the grand mean of scores for all participant and compares this against the error remaining in the model, which is the difference between the actual scores and the respective means of the groups. Therefore, the F-test is the ratio of systematic variance : unsystematic variance, so higher scores are better.

Going a little beyond, an ANOVA only tells you that there are differences between your groups, not where they lie. Therefore you would use a priori contrasts to test hypothesised differences between the groups or post-hoc tests to examine where the differences lie.

* For one variable case (testing two samples), t^2 = F, the square of t-statistic is equal to value of F-statistic.

Note: If the number of groups is k, the number of comparisons that can be done is: n=(k*(k-1)) / 2. If k=2 (pairwise comparison), n=1. If k=3, n=3. If k=4, n=6 and so forth.

Note: Familywise error = 1 - (0.95)^n, where n is the number of hypothesis tests performed. If doing pairwise test, there is 5% pct chance of false positive. If making three n-tests (3 groups), there is 14.3% chance of finding a false positive.

Think the F-ratio as a "signal-to-noise", where the signal is the error/change caused by the hypothesis model, and the noise is the background noise of the data. The stronger(higher) F-ratio, the more we can trust that our hypothesis model, is not due to random noise.

## How to calculate the p-value

If k=3, one would perform an ANOVA test between the 3 groups. Let's say that in each of these groups, we have 5 measurements (a balanced design). 
Our null-hypothesis H0, is there is no difference between the 3 groups. 
The other hypothesis is, that there is a difference between the three groups, due to a treatment. 
In most normal cases, we would like to "disprove/reject" the "null-hypothesis", to say that there is a difference between the three groups, due to a treatment. And that this difference is not due to random error. 
How is it done ? 
One would first determine the F-ratio to determine if there is any difference between the 3 groups (we still don't know/care which group differ). 
Let's say this F value is 5.12. This "signal-to-noise" is rather high. From this number we can expect that there is a variation between the 3 groups means, which is not due to random errors.

The between groups of freedom (dof_g) is equal: dof_g = 3 -1 
The within-group degrees of freedom (dof_w) is: dof_w= (5-1)+(5-1)+(5-1) 
The -1 is because that we for each group find an average, and "use one degree of freedom on this".

One can from the F-distribution calculate the probability for observing the F-ratio of 5.12. 
One would report these numbers: 
* F(dof_g, dof_w) = F(2, 12)
* p = 0.025

To calculate p-value in python do:

```python
from scipy import stats
p=stats.fprob(2, 12, 5.12)
print p
```

We can here state, that there is a significant difference between the groups. The p-value is lower than the "normal" 5 pct error. p < 0.05.

If one finds a significant difference, then afterwards one would test which group differ from each. These tests are normally called post-hoc tests.

Note: We can do the inverse of the probability to find the F-value given probability and the degrees of freedom.

```python
from scipy import special
F = special.fdtri(2, 12, 1-0.025)
print F
 
F = special.fdtri(2, 12, 1-0.0246762282557)
print F
```

## What is the meaning of p-value ?

The P value is a slippery concept, and the following statement is not true, but gives an idea "how-to-think". 
The p-value "can be interpreted" the probability that you wrongly reject" the "null-hypothesis". Or in another way, the p-value "can be interpreted" as the probability that you make a false positive. So, the lower p-value the better. 
To put it different 1 - p-value, "can be interpreted" as the probability that you can accept the other hypothesis, that there is difference. Or in another way, the 1 - p-value "can be interpreted" as the probability that you make a true positive.

The more correct interpretation is taken from: http://blog.minitab.com/blog/adventures-in-statistics/how-to-correctly-interpret-p-values 
One has to acknowledge/understand, that we in reality are performing tests on samples from a whole population of samples.

* A low p value suggests that your sample provides enough evidence that you can reject the null hypothesis for the entire population.

* A p value is the probability of obtaining an effect at least as extreme as the one in your sample data, assuming the truth of the null hypothesis.

* In another way. Expect the null hypothesis to be true. How often would an observed effect be due to sampling error? And how often would it not be due to sampling error.
* If p=0.04. If the null hypothesis is true, you would obtain the observed difference or more in 4% of studies due to random sampling error.
* If p=0.01. If the null hypothesis is true, you would obtain the observed difference or more in 1% of studies due to random sampling error.

Again, the lower p, the better. If we repeat the whole experiment 100 times with same sample size each time, there is 4% (or even only 1%), that the observed difference is due to random sampling error. The "rest" of the times, the observed difference is "not due" to random sampling error.

## Student's t-test

For data, see Student's t-test, Worked examples: https://en.wikipedia.org/wiki/Student%27s_t-test#Worked_examples

```python
import numpy as np
from scipy import stats
 
A1 = np.array([30.02, 29.99, 30.11, 29.97, 30.01, 29.99])
A2 = np.array([29.89, 29.93, 29.72, 29.98, 30.02, 29.98])
 
X1 = np.mean(A1)
X2 = np.mean(A2)
 
nominator = X1 - X2
print "X1 - X2 = %1.3f"%nominator
 
sd1 = np.std(A1, ddof=1)
sd2 = np.std(A2, ddof=1)
 
print "Sample standard deviations std_1=%1.2f and std_2=%1.2f"%(sd1, sd2)
 
# To square the standard deviation, you get the variance.
s1 = np.square(sd1)
n1 = len(A1)
 
s2 = np.square(sd2)
n2 = len(A2)
 
denominator = np.sqrt(s1/n1 + s2/n2 )
print "Denominator is : %1.4f"%(denominator)
 
df = np.square(s1/n1 + s2/n2) / ( np.square(s1/n1)/(n1-1) + np.square(s2/n2)/(n2-1) )
print "Degrees of freedom is : %1.2f"%(df)
 
t = nominator / denominator
 
print "The test statistic is : %1.3f"%(t)
 
p_one_tailed = stats.distributions.t.sf(np.abs(t), df)
p_two_tailed = p_one_tailed * 2
 
print 'The two-tailed test p-value is approximately %1.3f and the one-tailed p-value is approximately %1.3f'%(p_two_tailed, p_one_tailed)
 
df = 10
p_one_tailed = stats.distributions.t.sf(np.abs(t), df)
p_two_tailed = p_one_tailed * 2
print 'With df=10, the two-tailed test p-value is approximately %1.3f and the one-tailed p-value is approximately %1.3f'%(p_two_tailed, p_one_tailed)
```

## Balanced ANOVA F-test given complete datasets

For data, see F-test, One-way_ANOVA_example: https://en.wikipedia.org/wiki/F-test#One-way_ANOVA_example

```python
%matplotlib inline
import numpy as np
from scipy import stats
from scipy import special
import matplotlib.pyplot as plt
 
a1 = np.array([6, 8, 4, 5, 3, 4])
a2 = np.array([8, 12, 9, 11, 6, 8])
a3 = np.array([13, 9, 11, 8, 7, 12])
 
# Calculate the mean within each group:
Y1 = np.mean(a1)
Y2 = np.mean(a2)
Y3 = np.mean(a3)
 
print "The mean within each group: Y1, Y2, Y3 = %1.1f %1.1f %1.1f"% (Y1, Y2, Y3)
 
# Calculate the overall mean
Y = np.mean([Y1, Y2, Y3])
 
print "The overall mean: Y = %1.1f"% (Y)
 
# Calculate the "between-group" sum of squared differences:
n1 = len(a1)
n2 = len(a2)
n3 = len(a3)
S_B = n1*np.square(Y1-Y) + n2*np.square(Y2-Y) + n3*np.square(Y3-Y)
print "The 'between-group' sum of squared differences: S_B = %1.1f"% (S_B)
 
# The between-group degrees of freedom is one less than the number of groups
f_b = 3-1
 
# The between-group mean square
MS_B = S_B/f_b
print "The 'between-group' mean square: MS_B = %1.1f"% (MS_B)
 
# Centering the data in each group
a1_c = a1 - Y1
a2_c = a2 - Y2
a3_c = a3 - Y3
 
# The within-group sum of squares
S_W = np.sum( np.square(a1_c) + np.square(a2_c) + np.square(a3_c) )
print "The 'within-group' sum of squares: S_W = %1.1f"% (S_W)
 
# The within-group degrees of freedom
f_W = (n1-1)+(n2-1)+(n3-1)
print "The 'within-group' degrees of freedom: f_W = %i"% (f_W)
 
# The within-group mean square value is
MS_W = S_W / f_W
print "The 'within-group' mean square value is: MS_W = %1.1f"% (MS_W)
 
# The F-Ratio
F = MS_B/MS_W
 
print "The 'F-ratio' is: F = %1.1f"% (F)
 
# The critical F-value
alpha = 0.05
pct = 1.0 - alpha
 
# fdtr       -- Integral from 0 to x of F pdf.
# fdtrc      -- Integral from x to infinity under F pdf.
# fdtri      -- Inverse of fdtrc
 
Fcrit = special.fdtri(f_b, f_W, pct)
 
print "The critical F-value is: Fcrit = %1.1f"% (Fcrit)
 
print "Is F>Fcrit at alpha=%1.2f? = %s"% (alpha, F>Fcrit)
 
# The p-value for this tes
p = stats.fprob(f_b, f_W, F)
print "The p-value for this test: p= %1.3f"% (p)
 
# Comparison with library function
F_value, p_value = stats.f_oneway(a1, a2, a3)
 
print "scipy.stats.f_oneway library function. F_value=%3.3f, p_value=%3.3f"%(F_value, p_value)
 
# Make the plot
x_arr = np.linspace(0.1, 11, num=50)
F_dens = []
for x in x_arr:
    F_dens.append(1-special.fdtr(f_b, f_W, x))
 
plt.plot(x_arr, F_dens)
plt.ylabel('F density')
 
x_point = 3.682
y_point = 1-special.fdtr(f_b, f_W, x_point)
 
plt.annotate('x=%3.3f, y=%3.3f'%(x_point, y_point), xy=(x_point, y_point), xytext=(x_point, y_point*5), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
import numpy as np
import numpy.lib.recfunctions
from scipy import stats
import pandas as pd
 
c_m = 69.9
c_sd = 3.56
c_n = 15
 
groups = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J', 'K']
means = [81.2, 82.3, 70.6, 83.9, 71.2, 80.9, 80.7, 82.3, 83.2, 82.1, 80.7]
SDs = [2.60, 3.35, 3.70, 3.86, 3.14, 2.05, 2.87, 4.03, 3.47, 2.31, 2.99]
ns = len(groups)*[15]
 
data = np.array([groups, means, SDs, ns])
data = np.core.records.fromarrays(data, names='group, mean, SD, n', formats = 'S1, f8, f8, i8')
 
# Try a t calculation with Student's t-test, Equal sample sizes, equal variance
# https://en.wikipedia.org/wiki/Student%27s_t-test#Equal_sample_sizes.2C_equal_variance
t_es_ev = np.abs( (c_m - data['mean'])/( np.sqrt( (c_sd**2+data['SD']**2) )*np.sqrt(1./c_n) ) )
data = np.lib.recfunctions.append_fields(data, 't_es_ev', t_es_ev, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# Try a t calculation with Student's t-test, Equal or unequal sample sizes, equal variance
# https://en.wikipedia.org/wiki/Student%27s_t-test#Equal_or_unequal_sample_sizes.2C_equal_variance
t_us_ev = np.abs( (c_m - data['mean'])/( np.sqrt( ((c_n-1)*c_sd**2+(data['n']-1)*data['SD']**2)/(c_n+data['n']-2) )*np.sqrt(1./c_n + 1./data['n']) ) )
data = np.lib.recfunctions.append_fields(data, 't_us_ev', t_us_ev, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# Try a t calculation with Welch's t test, Equal or unequal sample sizes, unequal variances
# https://en.wikipedia.org/wiki/Welch%27s_t_test#Calculations
t_w = np.abs( (c_m - data['mean'])/np.sqrt(c_sd**2/c_n + data['SD']**2/data['n']) )
data = np.lib.recfunctions.append_fields(data, 't_w', t_w, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# The within-group degrees of freedom with students t-test
df_st = (c_n - 1) + (data['n'] - 1)
data = np.lib.recfunctions.append_fields(data, 'df_st', df_st, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# The p-value for this students t-test.
# The multiplication by 2, is the meaning of a two-tailed test
p_st = stats.distributions.t.sf(np.abs(data['t_es_ev']), df_st )*2
data = np.lib.recfunctions.append_fields(data, 'p_st', p_st, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# Test if above 0.05
ab_5p_st = p_st > 0.05
data = np.lib.recfunctions.append_fields(data, 'ab_5p_st', ab_5p_st, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# Try using the library function for probablity using F distribution
# The between-group degrees of freedom is one less than the number of groups
df_b = 2-1
 
# For one variable case: t^2 = F
F  = data['t_es_ev']**2
p_sf = stats.fprob(df_b, df_st, F)
data = np.lib.recfunctions.append_fields(data, 'p_sf', p_sf, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# The within-group degrees of freedom with Welch t-test
df_wt = np.square(c_sd**2/c_n + data['SD']**2/data['n']) / ( np.square(c_sd**2)/(np.square(c_n)*(c_n-1)) + np.square(data['SD']**2)/(np.square(data['n'])*(data['n']-1)) )
data = np.lib.recfunctions.append_fields(data, 'df_wt', df_wt, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
# The p-value for this Welch t-test
# The multiplication by 2, is the meaning of a two-tailed test
p_wt = stats.distributions.t.sf(np.abs(data['t_w']), df_wt )*2
data = np.lib.recfunctions.append_fields(data, 'p_wt', p_wt, dtypes=data['mean'].dtype, usemask=False, asrecarray=True)
 
## Use pandas to print pretty
data_pd = pd.DataFrame(data)
 
pd.options.display.float_format = '{:5,.3f}'.format
print data_pd
 
# Now sort the data according to the p-values
data_pd_s = data_pd.sort(['p_st'])
 
print "\nSorting according to Student tests p-value column."
print data_pd_s
```

## Difference Between Paired Means

This is taken from: http://stattrek.com/hypothesis-test/paired-means.aspx?Tutorial=AP

This is the situation, where you measure the same sample before and after treatment. 
Note, you do use the variance/uncertainty in this analysis.

Note:

The sampling distribution of the mean difference between data pairs (d) is approximately normally distributed.
Here's a simple rule to determine if the paired t must not be used - if a given data point in group one could be paired with any data point in group two, you cannot use a paired t test
See also:

* http://stattrek.com/estimation/mean-difference-pairs.aspx
* http://www.ruf.rice.edu/~bioslabs/tools/stats/ttest.html
*  http://www.tufts.edu/~gdallal/paired.htm

```python
# Python imports
import numpy as np
import numpy.lib.recfunctions
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
#from matplotlib import style
#style.use('ggplot')
 
before = [95, 89, 76, 92, 91, 53, 67, 88, 75, 85, 90, 85, 87, 85, 85, 68, 81, 84, 71, 46, 75, 80]
after = [90, 85, 73, 90, 90, 53, 68, 90, 78, 89, 95, 83, 83, 83, 82, 65, 79, 83, 60, 47, 77, 83]
 
# Create data array
data = np.array([before, after])
data = np.core.records.fromarrays(data, names='Training	, No_Training', formats = 'f8, f8')
 
# Calculate difference
d = data['Training'] - data['No_Training']
data = np.lib.recfunctions.append_fields(data, 'd', d, dtypes=data['Training'].dtype, usemask=False, asrecarray=True)
 
# Calculate the average of the differences
d_avg = np.average(data['d'])
# This is equal the difference between the mean of each group
X1_minus_X2 = np.average(data['Training']) - np.average(data['No_Training'])
 
print "Is the mean of differences equal ?", d_avg == X1_minus_X2, d_avg
 
# Calculate the standard deviation of the differences
SD_d = np.std(data['d'], ddof=1)
n = len(d)
print "The summed variance of differences are", SD_d**2*(n-1)
print "The standard deviation of the differences are", SD_d
 
# Test that: The population distribution of paired differences (i.e., the variable d) is normal.
nbins, bins, patches = plt.hist(data['d'], normed=True, histtype='stepfilled')
plt.setp(patches, 'facecolor', 'g', 'alpha', 0.75)
# add a line showing the expected distribution
y = mlab.normpdf( bins, d_avg, SD_d)
plt.plot(bins, y, 'k--', linewidth=1.5)
 
# Compute the standard error (SE) of the sampling distribution of d.
SE = SD_d / np.sqrt(n)
print "The standard error (SE) of the sampling distribution of d.", SE 
 
# Degrees of freedom. The degrees of freedom (DF) is: DF = n - 1 .
dof = n - 1
print "The degree of fredom is:", dof 
 
# Test statistic. The test statistic is a t-score (t) defined by the following equation.
# D is the hypothesized difference between population means, and SE is the standard error.
D = 0
t = (d_avg - D) / SE
print "The tests statistics are t=", t 
 
p_one_tailed = stats.distributions.t.sf(np.abs(t), dof)
p_two_tailed = p_one_tailed * 2
 
print 'The two-tailed test p-value is approximately %1.3f and the one-tailed p-value is approximately %1.3f'%(p_two_tailed, p_one_tailed)
 
alpha = 0.05
test = p_two_tailed > alpha
if test:
    print "Since the P-value (%1.3f) is greater than the significance level (%1.3f), we cannot reject the null hypothesis."%(p_two_tailed, alpha)
 
# Print by pandas
data_pd = pd.DataFrame(data)
pd.set_option('display.width', 1000)
#pd.options.display.float_format = '{:5,.4f}'.format
 
print data_pd
 
# Make a correlation plot
plt.figure()
plt.plot(data['Training'], data['No_Training'], linestyle='', marker='o', fillstyle='full')
# Ref line
x = np.linspace(data['Training'].min(), data['Training'].max(), num=50)
plt.plot(x, x, linestyle='-', marker='')
 
# Make confidence interval
confidence = 0.90
two_tailed_alpha = 1.00 - confidence
one_tailed_alpha = two_tailed_alpha/2
t_crit =  stats.distributions.t.isf(one_tailed_alpha, dof)
print 'The critical t value for confidence=%1.2f is equal:%1.3f'%(confidence, t_crit)
d_avg_min = d_avg + t_crit * SE
d_avg_max = d_avg - t_crit * SE
print "The %1.2f confidence interval of the average deviation %3.2f to %3.2f" %(confidence, d_avg_min, d_avg_max) 
 
# Make a deviation plot
plt.figure()
plt.plot(range(n), data['d'], linestyle='', marker='o', fillstyle='full')
plt.plot(range(n), np.ones(n)*d_avg, linestyle='-', marker='', fillstyle='full')
plt.plot(range(n), np.ones(n)*d_avg_min, linestyle='--', marker='', fillstyle='full')
plt.plot(range(n), np.ones(n)*d_avg_max, linestyle='--', marker='', fillstyle='full')
plt.title("Sample index of differences")
```

A repeated measures ANOVA is also referred to as a within-subjects ANOVA or ANOVA for correlated samples.

Data from: https://statistics.laerd.com/statistical-guides/repeated-measures-anova-statistical-guide.php

```python
import numpy as np
import numpy.lib.recfunctions 
from scipy import special
from scipy import stats
import pandas as pd
 
 
pre = [45, 42, 36, 39, 51, 44]
three = [50, 42, 41, 35, 55, 49]
six =[55, 45, 43, 40, 59, 56]
 
# Create data table
data = np.array([pre, three, six])
data = np.core.records.fromarrays(data, names='pre, three, six', formats = 'f8, f8, f8')
data_v = data.view((data.dtype[0], len(data.dtype.names)))
 
# Calculate the average across colums
col_means = np.average(data_v, axis=1)
 
# Calculate the average down trough rows.
row_means = np.average(data_v, axis=0)
 
# Grand mean
G_Mean = np.average(row_means)
print "Grand mean is:", G_Mean
 
# 
ni = len(data_v)
SStime = ni * np.sum( np.square( row_means - G_Mean ) )
print "SStime is :", SStime
 
SSw = 0.0
for i, name in enumerate(data.dtype.names):
    col_sum = np.sum( np.square( data[name] - row_means[i] ) )
    SSw += col_sum
print "SSw is :", SSw
 
k = len(data.dtype.names)
SSsubjects = k * np.sum( np.square(col_means - G_Mean) )
print "SSsubjects is :", SSsubjects
 
SSerror = SSw - SSsubjects
print "SSerror is :", SSerror
 
MStime = SStime/(k - 1)
print "MStime is :", MStime
 
MSerror = SSerror/( (ni - 1) * (k - 1) )
print "MSerror is :", MSerror
 
F = MStime / MSerror
print "F is :", F
 
# The critical F-value
# The between-group degrees of freedom is one less than the number of groups
f_b = k-1
# The within-group degrees of freedom
f_W = f_b*(ni-1)
 
alpha = 0.05
pct = 1.0 - alpha
 
Fcrit = special.fdtri(k - 1, ni - 1, pct)
print "Fcrit (%i, %i) at alpha=%3.2f is :%3.3f"%(f_b, f_W, alpha, Fcrit)
 
# The p-value for this tes
p = stats.fprob(f_b, f_W, F)
print "The p-value for this test: p= %1.3f"% (p)
 
# Print by pandas
# First add to table
data = np.lib.recfunctions.append_fields(data, 'means', col_means, dtypes=data['pre'].dtype, usemask=False, asrecarray=True)
 
# Add 0.0 in the end of the array.
row_means_row = np.append(row_means, [0.0])
# Resize and add to bottom
data = np.resize(data, (data.shape[0]+1) )
data[-1] = row_means_row
 
data_pd = pd.DataFrame(data)
pd.set_option('display.width', 1000)
#pd.options.display.float_format = '{:5,.4f}'.format
 
print data_pd
```

## Wilcoxon signed-rank test for matched pairs

See

* https://en.wikipedia.org/wiki/Wilcoxon_signed-rank_test
For dependent samples when the population cannot be assumed to be normally distributed

Try this: http://vassarstats.net/wilcoxon.html 
n=10

```python
from scipy import stats
import numpy as np
import numpy.lib.recfunctions
import pandas as pd
 
x1 = [125, 115, 130, 140, 140, 115, 140, 125, 140, 135]
x2 = [110, 122, 125, 120, 140, 124, 123, 137, 135, 145] 
 
# T and p value, two-sided
#p-value : The two-sided p-value for the test.
# T : The sum of the ranks of the differences above or below zero, whichever is smaller.
 
T, p = stats.wilcoxon(x1, x2)
print "Standard. T and p: ", T, p
 
# apply continuity correction by adjusting the Wilcoxon rank statistic by 0.5 towards the mean value when computing the z-statistic.
T, p = stats.wilcoxon(x1, x2, correction=True)
print "Correction. T and p: ", T, p
 
# Calculate the difference
diff = np.asarray(x1) -  np.asarray(x2)
 
# Select only those who passes the different from zero test
diff_c = np.compress(np.not_equal(diff , 0), diff )
x1_c = np.compress(np.not_equal(diff , 0), x1 )
x2_c = np.compress(np.not_equal(diff , 0), x2 )
 
## Do it manually
data = np.array([x1_c, x2_c])
data = np.core.records.fromarrays(data, names='x1, x2', formats = 'f8, f8')
 
data = np.lib.recfunctions.append_fields(data, 'sgn', np.sign(diff_c), float, usemask=False, asrecarray=True)
data = np.lib.recfunctions.append_fields(data, 'abs', np.abs(diff_c), float, usemask=False, asrecarray=True)
 
## Order by absolute difference
data.sort(order='abs')
 
## Rank the data
rank = stats.rankdata(data['abs'])
data = np.lib.recfunctions.append_fields(data, 'Ri', rank, float, usemask=False, asrecarray=True)
 
sgn_ri = data['sgn'] * data['Ri']
data = np.lib.recfunctions.append_fields(data, 'sgn_Ri', sgn_ri, float, usemask=False, asrecarray=True)
 
# Number of ranks
Nr = len(rank)
print "Nr is:", Nr
 
W = np.sum(sgn_ri)
print "W is:", W
 
STD_w = np.sqrt( (Nr*(Nr+1)*(2*Nr+1))/(6) )
print "STD_w is:", STD_w
 
z = (W - 0.5)/STD_w
print "z is:", z
 
# Confidence
confidence = 0.95
two_tailed_alpha = 1.00 - confidence
one_tailed_alpha = two_tailed_alpha/2
 
z_crit = stats.distributions.norm.isf(one_tailed_alpha )
print "z_crit=%1.3f for one-tailed alpha=%1.3f"%(z_crit, one_tailed_alpha)
 
# Probability
p_one_tailed = stats.distributions.norm.sf(z)
p_two_tailed = p_one_tailed * 2
print 'The two-tailed test p-value is approximately %1.3f and the one-tailed p-value is approximately %1.3f'%(p_two_tailed, p_one_tailed)
 
 
## Print by pandas
data_pd = pd.DataFrame(data)
pd.set_option('display.width', 1000)
pd.options.display.float_format = '{:5,.4f}'.format
print data_pd
```
